persistance gives us these 4 advantages:-

1) Short term memory to keep things in context 

2) Fault tolerance as in workflow5 while executing workflow.invoke(initial_state,config=config),if we stop it and in next cell again call it using workflow.invoke(none,config=config) ,the nodes that are already ran will not run again and it will continue from ahead of them.
3) Human In the Loop:-Example  Linkedin Post generator
          Generate Post -> Linkedin Post -> (Take permisiion before posting step) -> API call to linkedin

          ** this permission can take any amount of time
    So langgraph to wait for the permission suspends the workflow through interupt ,
    and when langgraph get the permission it resumes the flow using the previous persistent in memory 
4) Time Travel :- Can replay a workflow, means if a workflow execution is done ,i can go back to a previous node and start workflow execution from there
   -- we get do
          ** workflow.get_state_history(config1) and find the checkpoint id of the node from where we want the execution to begin from 
   -- then we get ** workflow.get_state({"configurable": {"thread_id":1, "checkpoint_id":"234i7sd"}})  to check out the state at this point of time
   -- then we do ** workflow.invoke(none,{"configurable": {"thread_id":1, "checkpoint_id":"234i7sd"}} )
